 # Acerca de la empresa

En 2016, Cyclistic lanzó una exitosa oferta de bicicletas compartidas. Desde entonces, el programa creció hasta alcanzar una flota de 5,824 bicicletas georreferenciadas y bloqueadas en una red de 692 estaciones en toda Chicago. Las bicicletas se pueden desbloquear desde una estación y devolverse en cualquier otra estación del sistema en cualquier momento.

Hasta ahora, la estrategia de marketing de Cyclistic se basaba en la construcción de un reconocimiento de marca general y en atraer a amplios segmentos de consumidores. Uno de los enfoques que ayudó a hacer esto posible fue la flexibilidad de sus planes de precios: pases de un solo viaje, pases de un día completo y membresías anuales. A los clientes que compran pases de un solo viaje o pases de un día completo se los llama ciclistas ocasionales. Los clientes que compran membresías anuales se llaman miembros de Cyclistic.

Los analistas financieros de Cyclistic llegaron a la conclusión de que los miembros anuales son mucho más rentables que los ciclistas ocasionales. Aunque la flexibilidad de precios ayuda a Cyclistic a atraer más clientes, Moreno cree que maximizar el número de miembros anuales será clave para el crecimiento futuro. En lugar de crear una campaña de marketing que apunte a todos los clientes nuevos, Moreno cree que hay muchas posibilidades de convertir a los ciclistas ocasionales en miembros. Ella señala que los ciclistas ocasionales ya conocen el programa de Cyclistic y han elegido a Cyclistic para sus necesidades de movilidad.

Podemos observar que existen valores nulos en las columnas 'start_station_name', 'start_station_id', 'end_station_name', 'end_station_id', 'end_lat' y 'end_lng', pero para nuestro análisis no vamos a trabajar con estas columnas por lo que no representan datos críticos para el análisis así que procedemos.

A continuación unimos todos nuestros conjuntos de datos para concentrarlos solo en uno, después elimino las columnas que no vamos requerir para el análisis, enseguida agrego una columna con el tiempo de duración de cada viaje y elimino registros erróneos, como aquellos que tienen una duración de tiempo negativa y con una duración igual a 0.

```python
tripdata_2022=[tripdata_202201,tripdata_202202,tripdata_202203,tripdata_202204,tripdata_202205,tripdata_202206,tripdata_202207,tripdata_202208,tripdata_202209,tripdata_202210,tripdata_202211,tripdata_202212]
tripdata_2022=pd.concat(tripdata_2022)

tripdata_2022=tripdata_2022.drop(columns=['start_station_name','start_station_id','end_station_name',
                                         'end_station_id','start_lat','start_lng','end_lat','end_lng'])

tripdata_2022['started_at']=tripdata_2022['started_at'].astype('datetime64')
tripdata_2022['ended_at']=tripdata_2022['ended_at'].astype('datetime64')
tripdata_2022['ride_length']=round((tripdata_2022['ended_at']-tripdata_2022['started_at'])/pd.Timedelta(minutes=1),2)
tripdata_2022

tripdata_2022.reset_index(inplace=True)
del tripdata_2022['index']
tripdata_2022

indice=[]
for i,e in enumerate(tripdata_2022['ride_length']):
    if e<0 or e == 0:
        indice.append(i)
tripdata_2022.drop(indice,axis=0,inplace=True)
tripdata_2022
```

# 4. Analizar (Analyze)

La siguiente fase del proceso corresponde al análisis de los datos. En esta fase visualizaremos la distribución que tienen ciertos campos, desarrollaremos algunas medidas críticas para después analizarlas mediante visualizaciones gráficas en Microsoft Power BI.

A continuación, generamos un par de gráficos para observar como se distribuyen los datos de la columna 'ride_length_min' en función del tipo de usuario y del tipo de bicicleta.

```python
fi, ax = plt.subplots(nrows=1,ncols=2,figsize=(12,8),sharex=False)

plt.subplot(1,2,1)
ax1=sns.boxplot(data=tripdata_2022,x='member_casual',y='ride_length')

plt.subplot(1,2,2)
ax2=sns.boxplot(data=tripdata_2022,x='rideable_type',y='ride_length')

plt.show()
```

Notamos que los valores máximos superan los 40000 minutos de duración de un viaje, así que decido eliminar el 1% de los valores máximos y el 1% de los valores mínimos también, para así quedarnos con el 98% de los registros del conjunto de datos y procedemos a volver a graficar y observar su distribución.

```python
percent_99=np.percentile(tripdata_2022['ride_length'],99)
percent_01=np.percentile(tripdata_2022['ride_length'],1)
tripdata_2022_98=tripdata_2022.loc[(tripdata_2022['ride_length'] <= percent_99) & (tripdata_2022['ride_length']>=percent_01)]

fi, ax = plt.subplots(nrows=1,ncols=2,figsize=(12,8),sharex=False)

plt.subplot(1,2,1)
ax1=sns.boxplot(data=tripdata_2022_98,x='member_casual',y='ride_length')

plt.subplot(1,2,2)
ax2=sns.boxplot(data=tripdata_2022_98,x='rideable_type',y='ride_length')

plt.show()
```

Y ahora vemos como los valores máximos superan por poco los 100 minutos de duración del viaje y como se concentran la mayoría de los viajes en una duración dentro de los 10 y 30 minutos.
